```python
import os
import glob
import utmosv2
import numpy as np

model = utmosv2.create_model(pretrained=True)

# Directory containing audio files
folder = "/content/drive/MyDrive/266/eval/hfft"

audio_files = sorted(glob.glob(os.path.join(folder, "*.wav")))

mos_scores = []

print("Evaluating MOS for files...\n")

for path in audio_files:
    mos = model.predict(input_path=path)
    mos_scores.append(float(mos))

    print(f"{os.path.basename(path)} → MOS: {mos}")

mean_mos = np.mean(mos_scores)
print("\n==============================")
print(f"Mean MOS for folder: {mean_mos:.4f}")
print("==============================")
```

```python
import whisper
import pandas as pd
import random
from jiwer import wer

meta_path = "/content/drive/MyDrive/266/datasplit/test_metadata.csv"
df = pd.read_csv(meta_path)

df_sample = df.sample(n=100, random_state=42)

model = whisper.load_model("small")

wers = []

print("Evaluating WER on 100 random test samples...\n")

for idx, row in df_sample.iterrows():
    audio_path = row["file_path"]
    reference = row["normalized_transcript"]

    # Whisper transcription
    result = model.transcribe(audio_path)
    hypothesis = result["text"]

    score = wer(reference.lower(), hypothesis.lower())
    wers.append(score)

    print(f"{audio_path.split('/')[-1]} → WER: {score:.4f}")

mean_wer = sum(wers) / len(wers)

print("\n======================================")
print(f"Mean WER over 100 samples: {mean_wer:.4f}")
print("======================================")
```
